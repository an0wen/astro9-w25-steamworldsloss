```python
import numpy as np
import matplotlib.pyplot as plt
import iop_ext as iop

print("Modules imported successfully")
```

```python
# import the masses and radii of exoplanets catalog
# we will see another time how we actually got this
# the data is imported using numpy genfromtxt, meaning that data is imported in numpy arrays
# explain each option and what it is used for
# NEA stands for NASA Exoplanet Archive
nea_output_file = "./data/catalog_exoplanets.dat"
list_catalog_rp,list_catalog_mp \
    = np.genfromtxt(nea_output_file,delimiter="\t",unpack=True,usecols=(1,4),filling_values=0.0)
print("Exoplanet catalog imported successfully")
```

```python
# import mass-radius models from Li Zeng 2016
# explain each option and what it is used for
# genfromtext is a little more elaborate (and slower) than loadtxt, but they do essentially the same thing
list_zeng_fe_m,list_zeng_fe_r = np.loadtxt("./models/zeng-iron.dat",unpack=True,usecols=(0,1))
list_zeng_ea_m,list_zeng_ea_r = np.loadtxt("./models/zeng-earth.dat",unpack=True,usecols=(0,1))
list_zeng_mg_m,list_zeng_mg_r = np.loadtxt("./models/zeng-rock.dat",unpack=True,usecols=(0,1))
print("Mass-radius data imported successfully")
```

```python
# making the plot
# explain difference between line and scatter (they probably did it already in class, so just briefly remind them)

N_exo = len(list_catalog_mp)
point_sizes = np.ones(N_exo)*5.0

# plotting the exoplanets
plt.scatter(list_catalog_mp,list_catalog_rp,color='black',alpha=0.5,sizes=point_sizes)

# plot the models
plt.plot(list_zeng_fe_m,list_zeng_fe_r,color="black",linestyle='-',linewidth=2)
plt.plot(list_zeng_ea_m,list_zeng_ea_r,color="green",linestyle='-',linewidth=2)
plt.plot(list_zeng_mg_m,list_zeng_mg_r,color="brown",linestyle='-',linewidth=2)

# adding some options to make the figure prettier, and also focus on important stuff
plt.xscale('log')
plt.xlim(0.2,50)
plt.ylim(0.5,4.5)
```
